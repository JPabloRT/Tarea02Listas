```
#Problema 3: Escribir un programa que cuente la frecuencia de los elementos de una lista. 

lista = [9, 7, 1, 6, 9, 14, 25 , 15, 16, 8, 11, 6, 8, 4, 9]
print("Teniendo la siguiente lista: " + str(lista))
x = input("Ingrese el número a contar: ")
conteo = lista.count(int(x))

print("El número que ingreso para contar es " + str(x) + ", este número aparece " + str(conteo) + " en la lista.")
```
